```python
import io
import os
import json
from tqdm import tqdm

def store_fasttext_embeddings(fname, embedding_dir, lang):
    fin = io.open(
        os.path.join(embedding_dir, fname),'r', encoding='utf-8', newline='\n', errors='ignore'
        )
    n, d = map(int, fin.readline().split()) 
    data = {}
    for line in tqdm(fin, total = n):
        tokens = line.rstrip().split(' ')
        data[tokens[0]] = list(map(float, tokens[1:])) 
    return data
```

```python
# 15:24

data = store_fasttext_embeddings("cc.hi.300.vec", "fasttext", "hi")
```
